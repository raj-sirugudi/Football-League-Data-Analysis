# Matches by League

```python
#Matches by league

plt.figure(figsize=(8,4))
fx = sns.countplot(y = Football_League_alldata["Date"],order=Football_League_alldata["Date"].value_counts().index, linewidth = 2,
edgecolor = "K"*Football_League_alldata["Date"].nunique())
for i,j in enumerate(Football_League_alldata["Date"].value_counts().values):
    fx.text(.8,i,j,weight = "bold")
plt.title("Matches by league from 2012/2015")
plt.show()

#Below Figure suggests that all the four seasons list is of same format and had equal number of games.
```

# Home and Away team goals by league



```python
#FTHG=Football Home Team goals
#FTAG=Football Away Team goals

Football_League_alldata.groupby("Date").agg({"FTHG":"sum","FTAG":"sum"}).plot(kind="barh",figsize = (10,10),edgecolor = "k",
linewidth =1)
plt.title("Home Team and Away Team goals by league")
plt.legend(loc = "best" , prop = {"size" : 18})
plt.xlabel("total goals")
plt.show()
```

# Distribution of Home and Away Goals

```python
plt.figure(figsize=(12,6))
sns.kdeplot(Football_League_alldata["FTHG"],shade=True,
            color="b",label="home goals")
sns.kdeplot(Football_League_alldata["FTAG"],shade=True,
            color="r",label="away goals")
plt.axvline(Football_League_alldata["FTHG"].mean(),linestyle = "dashed",
            color="b",label="home goals mean")
plt.axvline(Football_League_alldata["FTAG"].mean(),linestyle = "dashed",
            color="r",label="away goals mean")
plt.legend(loc="best",prop = {"size" : 12})
plt.title("DISTRIBUTION OF HOME AND AWAY GOALS")
plt.xlabel("goals")
plt.show()
```

# Proportion of game outcomes

```python
plt.figure(figsize=(8,8))
Football_League_alldata["FTR"].value_counts().plot.pie(autopct = "%1.0f%%",colors =sns.color_palette("rainbow",3),wedgeprops = {"linewidth":2,"edgecolor":"gray"})
Football_circ = plt.Circle((0,0),.7,color = "gray")
plt.gca().add_artist(Football_circ)
plt.title("PROPORTION OF GAME OUTCOMES")
plt.show()

# FTR is Football match result
# H is Home team wins
# A is Away team wins
# D is draw
```

# Creating new column for Total number of goals in a match

```python
Football_League_alldata["Total_Goals"]=Football_League_alldata["FTHG"]+Football_League_alldata["FTAG"]
#FTHG Home team goals
#FTAG Away team goals
```

# Goals Scored in League 

```python
plt.figure(figsize=(13,5))
sns.violinplot(Football_League_alldata["Date"],Football_League_alldata["Total_Goals"],palette="rainbow")
plt.title("GOALS SCORED IN LEAGUES")
plt.xticks(rotation = 60)
plt.show()
```

# Summing up fouls,yellow cards, red cards for both Home and Away teams

```python
Football_League_alldata["Home_Fouls_Cards"]=Football_League_alldata["HF"]+Football_League_alldata["HY"]+Football_League_alldata["HR"]
#HF HomeTeam Fouls
#HY HomeTeam Yellow Cards
#HR HomeTeam Red Cards
```

```python
Football_League_alldata["Away_Fouls_Cards"]=Football_League_alldata["AF"]+Football_League_alldata["AY"]+Football_League_alldata["AR"]
#AF AwayTeam Fouls
#AY AwayTeam Yellow Cards
#AR AwayTeam Red Cards
```

```python
Football_League_alldata.groupby("Date").agg({"Home_Fouls_Cards":"sum","Away_Fouls_Cards":"sum"}).plot(kind="barh",figsize = (10,10),edgecolor = "k",
linewidth =1)
plt.title("Home Team and Away Team Fouls and cards")
plt.legend(loc = "best" , prop = {"size" : 18})
plt.xlabel("Total Fouls")
plt.show()
```

# Home and Away teams Corner kicks and shots on target comparison

```python
Football_League_alldata["Home_Kicks"]=Football_League_alldata["HST"]+Football_League_alldata["HC"]
#HST HomeTeam strikes on target
#HC HomeTeam Corner kicks
```

```python
Football_League_alldata["Away_Kicks"]=Football_League_alldata["AST"]+Football_League_alldata["AC"]
#AST AwayTeam strikes on target
#AC AwayTeam Corner kicks
```

```python
Football_League_alldata.groupby("Date").agg({"Home_Kicks":"sum","Away_Kicks":"sum"}).plot(kind="barh",figsize = (8,8),edgecolor = "k",
linewidth =1)
plt.title("Home Team and Away Team Corner kicks and shots on target")
plt.legend(loc = "best" , prop = {"size" : 18})
plt.xlabel("Total Kicks")
plt.show()
```

```python
home_t = Football_League_alldata.groupby("HomeTeam")["FTHG"].sum().reset_index()
away_t = Football_League_alldata.groupby("AwayTeam")["FTAG"].sum().reset_index()
home_t = home_t.sort_values(by="FTHG",ascending= False)
away_t = away_t.sort_values(by="FTAG",ascending= False)
plt.figure(figsize=(13,10))
plt.subplot(121)
ax = sns.barplot(y="HomeTeam",x="FTHG",
                 data=home_t[:20],palette="winter",
                 linewidth = 1,edgecolor = "k"*20)
plt.ylabel('')
plt.title("Top teams by home goals")
for i,j in enumerate(home_t["FTHG"][:20]):
    ax.text(.7,i,j,weight = "bold")
plt.subplot(122)
ax = sns.barplot(y="AwayTeam",x="FTAG",
                 data=away_t[:20],palette="summer",
                linewidth = 1,edgecolor = "k"*20)
plt.ylabel("")
plt.subplots_adjust(wspace = .4)
plt.title("Top teams by away goals")
for i,j in enumerate(away_t["FTAG"][:20]):
    ax.text(.7,i,j,weight = "bold")
```
